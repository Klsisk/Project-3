```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in salary csv
salary_df = pd.read_csv("static/data/salary.csv")
salary_df.head(10)
```

```python
# Sort by ascending salaries
sort_salary = salary_df.sort_values('salary',ascending=False)
sort_salary.head(11)
```

```python
# Generate a bar plot showing top 10 player salaries and their teams
# Set x_axis, y_axis, tick locations, and colors for bars
y_axis = ['0', 'MEM', 'OKC', 'TOR','DET', 'BOS', 'DEN', 'CLE', 'GSW']
x_axis = ['0', '28530608', '28530608', '28703704', '29512900', '29727900', '31269231', '33285709', '34682550']
color= ["navy", "red", "navy", "red", "navy", "red", "navy", "red"]

# Show the graph
plt.barh(y_axis, x_axis, color=color, align="center")
plt.xticks(rotation="45", fontsize=7)

# Set a Title and labels
plt.title('Top 8 Player Salaries and Their Teams')
plt.ylabel('Teams')
plt.xlabel('Player Salaries')

# Show the graph
plt.show()
```

```python
# Load in allstar csv
allstar_df = pd.read_csv("static/data/allstar1.csv")
allstar_df.head(10)
```

```python
total_teams = allstar_df["team"].value_counts()
total_teams
```

```python
# Generate a bar plot showing top 10 teams with most allstars
# Set x_axis, y_axis, tick locations, and colors for bars
x_axis = ['0', 'ORL','DET', 'ATL', 'PHX', 'DAL', 'HOU', 'SAS', 'BOS', 'LAL', 'MIA']
y_axis = ['0', '14', '15', '17', '18', '18', '19', '24', '25', '25', '28']
color= ["navy", "red", "navy", "red", "navy", "red", "navy", "red"]

plt.figure(figsize=(6,3))
plt.bar (x_axis,y_axis, color=color, align="center")
plt.xticks(rotation="45", fontsize=8)

# Set a Title and labels
plt.title('Top Teams With Most All-Stars')
plt.xlabel("Teams")
plt.ylabel("Number of Players")

# Show the graph
plt.show()
```
