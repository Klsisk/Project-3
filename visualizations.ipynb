```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in salary csv
salary_df = pd.read_csv("static/data/salary.csv")
salary_df.head()
```

```python
# Sort by ascending salaries
sort_salary = salary_df.sort_values('salary',ascending=False)
sort_salary.head(10)
```

```python
y_axis = ['0', 'DET', 'BOS', 'DEN', 'CLE', 'GSW']
x_axis = ['0', '29512900', '29727900', '31269231', '33285709', '34682550']

plt.barh(y_axis,x_axis)
plt.title('Top 5 Player Salaries and Their Teams')
plt.ylabel('Teams')
plt.xlabel('Player Salaries')
plt.show()
```
